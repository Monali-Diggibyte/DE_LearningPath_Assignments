```
df_emp = spark.read.format("csv").options(header = True, sep= ",", inferSchema = True).load("/FileStore/tables/employee_data-1.csv")

display(df_emp)
```

```
df_emp.printSchema()
```

```
from pyspark.sql.functions import *
```

```
df_emp_n = df_emp.select(col( "Employee_ID"), col( "Emp_First_Name"), col( "Emp_Last_Name"), col( "Experience"), col( "Salary"), \
                         col( "DOB"), to_date(col("DOB"), "dd-MM-yyyy").alias("Date of Birth"))
    
display(df_emp_n)
```

```
df_emp_n.show()
df_emp_n.printSchema()
```

```
df_emp_n.withColumn("Age in Months", months_between(current_date(),col("Date of Birth"))) \
        .withColumn("Age in Years", round(months_between(current_date(),col("Date of Birth"))/lit(12))) \
        .show()
```

```
df_emp_age = df_emp_n.withColumn("Age in Months", months_between(current_date(),col("Date of Birth"))) \
                     .withColumn("Age in Years", round(months_between(current_date(),col("Date of Birth"))/lit(12)))

display(df_emp_age)
df_emp_age.printSchema()
```
